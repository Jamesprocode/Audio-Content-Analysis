```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import audio file
audio, sr = librosa.load('/Users/jameswang/workspace/Audio Content Analysis/audios/vowels_sung.wav', sr=None)

# compute mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=128, fmax=8000)
# convert to log scale (dB)
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

# plot mel spectrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_mel_spectrogram, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.tight_layout()
plt.show()
```

```python
# Compute MFCC features from the audio signal
mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_fft=2048, n_mels=64, n_mfcc=13)
mfcc_maxnorm= (mfccs - np.max(np.abs(mfccs), axis=1, keepdims=True))/np.max(np.abs(mfccs), axis=1, keepdims=True)
# Plot the MFCC features
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc_maxnorm[1:], x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
